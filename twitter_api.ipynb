```python
import requests

base_url = 'https://api.twitter.com/'
auth_url = '{}oauth2/token'.format(base_url)

auth_headers = {
    'Authorization': 'Basic {}'.format(b64_encoded_key),
    'Content-Type': 'application/x-www-form-urlencoded;charset=UTF-8'
}

auth_data = {
    'grant_type': 'client_credentials'
}

auth_resp = requests.post(auth_url, headers=auth_headers, data=auth_data)
```

#### API.search(q [, lang] [, locale] [, rpp] [, page] [, since_id] [, geocode] [, show_user])

Returns tweets that match a specified query.

Parameters:	
* **q** â€“ the search query string
* **lang** â€“ Restricts tweets to the given language, given by an ISO 639-1 code.
* **locale** â€“ Specify the language of the query you are sending. This is intended for language-specific clients and the default should work in the majority of cases.
* **rpp** â€“ The number of tweets to return per page, up to a max of 100.
* **page** â€“ The page number (starting at 1) to return, up to a max of roughly 1500 results (based on rpp * page.
* **since_id** â€“ Returns only statuses with an ID greater than (that is, more recent than) the specified ID.
* **geocode** â€“ Returns tweets by users located within a given radius of the given latitude/longitude. The location is preferentially taking from the Geotagging API, but will fall back to their Twitter profile. The parameter value is specified by â€œlatitide,longitude,radiusâ€, where radius units must be specified as either â€œmiâ€ (miles) or â€œkmâ€ (kilometers). Note that you cannot use the near operator via the API to geocode arbitrary locations; however you can use this geocode parameter to search near geocodes directly.
* **show_user** â€“ When true, prepends â€œ<user>:â€ to the beginning of the tweet. This is useful for readers that do not display Atomâ€™s author field. The default is false.

Return type:	
**list of SearchResults objects**

```python
def user_information(username):
    information={}
    user=api.get_user(username)
    user_json=json.loads(json.dumps(user._json))
    username = user_json['screen_name']
    user_id = user_json['id']
    fecha = time.strptime(user_json['created_at'], '%a %b %d %H:%M:%S +0000 %Y')
    favoritos = user_json['favourites_count']
    followers = user_json['followers_count']
    following = user_json['friends_count']
    localizacion = str(user_json['location'].encode('utf-8'))
    information = {'user' : username,
                   'id_client' : user_id,
                   'date_creation' : time.strftime("%Y/%m/%d",fecha),
                   'number_favorites' : favoritos,
                   'number_followers' : followers,
                   'number_following' : following,
                   'location' : localizacion
                  }
    return information
```
